```python
# Import necessary libraries
import numpy as np
import pandas as pd

# For plotting graphs
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
raw_np = load_boston()
data = pd.DataFrame(raw_np.data, columns=raw_np.feature_names)
```

```python
data.head()
```

```python
target_arr = raw_np.target
target_data = pd.Series(target_arr, name='target')
target_data.head(2)
```

```python
data['target'] = target_data
data.head()
```

```python
corr = data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True, fmt='.2f', cmap="coolwarm")
```

```python
top_list = ["RM", "LSTAT", "TAX","RAD"]
col = ['r','b','g','c']

x_plot = data.loc[:,top_list]
y_plot = data['target']
c = 0

for x in x_plot:
    sns.regplot(x=x, y=y_plot, data=data, color=col[c])
    plt.show()
    c += 1
```

```python
import sklearn
```

```python
# sklearn.__all__
# model_selection.__all__
```

```python
from sklearn.model_selection import train_test_split
X = data.drop('target', axis=1)
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print("Shape of X_train is: ", X_train.shape)
print("Shape of X_test is: ", X_test.shape)
print("Shape of y_train is: ", y_train.shape)
print("Shape of y_test is: ", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_preds = model.predict(X_test)
```

```python
from sklearn import metrics
metrics.__all__
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
msr = mean_squared_error(y_test, y_preds)
mae = mean_absolute_error(y_test, y_preds)
r2s = r2_score(y_test, y_preds)
print(msr, " " , mae, " ", r2s)
```

```python
sns.regplot(x=y_test, y=y_preds, data=data)
plt.xlabel("actual price")
plt.ylabel("estimated price")
```
